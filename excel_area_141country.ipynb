```python
#20210508
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('E:\\BaP_Con(1).xlsx', skiprows = [0, 1, 2, 3], header=None)
df.head()
```

```python
#去掉无用的列
data = df.iloc[:, 4::] 
data.head()
```

```python
num = 141
x = 3
y = 7
a = np.zeros((num, num))
```

```python
for i in range(num):
    for j in range(num):
        a[j, i] = data.iloc[0+j*7:7*(j+1), 0 + i*x:3+i*3].sum(axis=1).sum()
```

```python
print(a) #a为141*141的数据
```

```python
#存为excel
data1 = pd.DataFrame(a) #转化为DataFrame格式
writer = pd.ExcelWriter('E:\\country141.xlsx') #存入的路径
data1.to_excel(writer)
writer.save()
```
